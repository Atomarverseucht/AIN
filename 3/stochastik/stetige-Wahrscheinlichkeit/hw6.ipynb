```python
import numpy as np
import scipy.stats as stat
```

```python
defekt = 0.25
D = stat.binom(1, defekt)
A = stat.binom(100, defekt)
W = stat.geom(defekt)
print(W.mean())
print(f"P(A <= 20) = {A.cdf(20)}")
print(f"P(W > 10) = {1 - W.cdf(10)}")
```

```python
H = stat.poisson(5)
print(f"P(telefon = 5) = {H.pmf(5)}")
print(f"P(telefon > 5) = {1 - H.cdf(5)}")
```

```python
B = stat.binom(10, 0.7)
E = stat.geom(0.3)
F = stat.poisson(40)
R = stat.poisson(20)

print(f"P(E = 5) = {E.pmf(5)}")
print(f"P(B >= 5) = {1 - B.cdf(4)}")
print(f"{B.ppf(0.95)}")
print(f"{F.cdf(30)}")
print(f"{R.cdf(30)}")
print(f"{stat.poisson(60).cdf(30)}")
```

```python
unten = 2
X = stat.uniform(unten, 4 - unten)
print(X.cdf(3))
print(X.cdf(3.5) - X.cdf(2.5))
print(X.mean())
print(X.var())
print(np.sqrt(X.var()))

print(X.median())
print(X.ppf(0.9))
```

```python

class ADist(stat.rv_continuous):
    def _pdf(self, x):
        return 1.5 * (x ** 2) + x

class BDist(stat.rv_continuous):
    def _pdf(self, x):
        return 3 * (x ** (-2))

A = ADist(a= 0, b= 1)
B = BDist(a=2,b= 6)
print(A.mean())
print(1 - B.cdf(4))
print(B.mean())
print(A.var())
```
