```python
import numpy as np
import scipy.stats as stats
```

```python
M = stats.binom(10, 0.45)
W = stats.geom(0.45)
print(f"E(M) = {M.mean()}")
print(f"E(W) = {W.mean()}")
print(f"P(M > 3) = {1 - M.cdf(3)}")
print({0.55 ** 4 * 0.45})
```

```python
A = stats.poisson(5)
B = np.array([0.1, 0.1, 0.2, 0.2, 0.4])
print(f"E(A) = {A.mean()}")
print(f"E(B) = {np.sum(B * range(1,6))}")
print(f"P(A = 5) = {A.pmf(5)}")
print(f"P(B >= 3) = {0.2 + 0.2 + 0.4}")
```
