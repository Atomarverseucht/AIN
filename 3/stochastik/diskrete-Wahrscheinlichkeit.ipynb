```python
import numpy as np
import scipy.stats as stats
```

```python
X = stats.binom(5, 0.25)
print(f"Wahrscheinlichkeit für genau 2: {X.pmf(2)}")
print(f"Wahrscheinlichkeit max. 2: {X.cdf(2)}")
print(X.pmf(range(0, X.args[0])))
```

```python
Disk = stats.binom(50, 0.03)
print(f"keine: {Disk.pmf(0):0.3f}")
print(f"genau 2: {Disk.pmf(2):0.3f}")
print(f"höchstens 2: {Disk.cdf(2):0.3f}")
print(f"Erwartungswert: {Disk.mean():0.3f}")
```

```python
Popcorn = stats.poisson(3)
print(f"keine: {Popcorn.pmf(0):0.3f}")
print(f"genau 2: {Popcorn.pmf(2):0.3f}")
print(f"höchstens 2: {Popcorn.cdf(2):0.3f}")
print(f"Erwartungswert: {Popcorn.mean():0.3f}")
```

```python
Coffee = stats.poisson(75)
print(f"max 70: {Coffee.cdf(70):0.3f}")
```
